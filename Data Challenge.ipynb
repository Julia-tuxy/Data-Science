```python
df = df[(df['age_group_name']=='Age standardized')]
```

```python
df = df[(df['location_name']!='Global')]
```

```python
df = df[(df['sex_name']!='Both')]
```

```python
df = df[(df['location_id']!=19)]
```

```python
locationgroup = df.groupby(['location_name','sex_name'])['val'].mean()
locationgroup
```

```python
head=locationgroup.sort_values(ascending=False).head(10)
head
```

```python
head=head.unstack(level='sex_name')
head
```

```python
locationgroup=locationgroup.unstack(level='sex_name')
```

```python
locationgroup.reset_index(level=0, inplace=True)
```

```python
locationgroup=locationgroup.rename(columns={'location_name' : 'country'})
locationgroup
```

```python
locationgroup = locationgroup[(locationgroup['country']!='Bolivia (Plurinational State of)')]
```

```python
import pycountry_convert as pc
```

```python
def convert(row):
    cn_code = pc.country_name_to_country_alpha2(row.country, cn_name_format = 'default')
    conti_code = pc.country_alpha2_to_continent_code(cn_code)
    return conti_code
```

```python
locationgroup['continet']=locationgroup.apply(convert,axis=1)
```

```python
locationgroup
```

```python
continent = locationgroup.groupby(['continet'])['Male'].mean()
continent
```

```python
continent = locationgroup.groupby(['continet'])['Female'].mean()
continent
```
